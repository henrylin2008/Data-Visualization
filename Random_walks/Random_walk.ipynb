```python
from random import choice


class RandomWalk:
    """A class to generate random walks."""

    def __init__(self, num_points=5000):    # default number of points in a walk
        """Initialize attributes of a walk."""
        self.num_points = num_points

        # All walks start at (0, 0).
        self.x_values = [0]
        self.y_values = [0]

    def fill_walk(self):
        """Calculate all the points in the walk."""

        # Keep taking steps until the walk reaches the desired length.
        while len(self.x_values) < self.num_points:

            # Decide which direction to go and how far to go in that direction.
            x_direction = choice([1, -1])   # 1 for right movement, -1 for left
            x_distance = choice([0, 1, 2, 3, 4])    # how far to move in that direction randomly between 0 and 4
            # length of each step in x direction; positive means move right, negative means move left
            # 0 means move vertically
            x_step = x_direction * x_distance

            y_direction = choice([1, -1])
            y_distance = choice([0, 1, 2, 3, 4])
            # length of each step in y direction; positive means move up, negative means move down
            # 0 means move horizontally
            y_step = y_direction * y_distance

            # Reject moves that go nowhere.
            if x_step == 0 and y_step == 0:
                continue

            # Calculate the new position.
            x = self.x_values[-1] + x_step  # next x-value
            y = self.y_values[-1] + y_step  # next y-value

            self.x_values.append(x)
            self.y_values.append(y)
```

```python
# plotting the random walk 
import matplotlib.pyplot as plt

# from random_walk import RandomWalk

# keep making new walks, as long as the program is active. 
while True:
    # Make a random walk.
    rw = RandomWalk(50_000)
    rw.fill_walk()

    # Plot the points in the walk.
    plt.style.use('classic')
    fig, ax = plt.subplots(figsize=(15, 9), dpi=128)  # figsize sets the size of the figure, dpi sets a plot size 
    point_numbers = range(rw.num_points)  # list of numbers = # of points in the walk 
    ax.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap=plt.cm.Blues, edgecolors='none', s=1)
    
    # Emphasize the first and last points.
    ax.scatter(0, 0, c='green', edgecolors='none', s=100)   # starting point, greeen, size: 100
    ax.scatter(rw.x_values[-1], rw.y_values[-1], c='red', edgecolors='none', s=100) # end point, red, size: 100

    # Remove the axes.
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
    plt.show()
    
    keep_running = input("Make another walk? (y/n): ")
    if keep_running == 'n':
        break
```
