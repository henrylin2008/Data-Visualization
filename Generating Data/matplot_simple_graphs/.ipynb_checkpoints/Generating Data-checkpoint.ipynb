```python
# mpl_squares
import matplotlib.pyplot as plt

input_values = [1, 2, 3, 4, 5]
squares = [1, 4, 9, 16, 25]

# print(plt.style.available)    # list of available style
plt.style.use('seaborn')
fig, ax = plt.subplots()
# overwrite the default behaviour of starting x-coordinate (0) with input_values
ax.plot(input_values, squares, linewidth=3)   # linewidth = thickness fo the line 

# set chart title and label axes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Set size of tick labels.
ax.tick_params(axis='both', labelsize=14)

plt.show()
```

```python
# Plotting a series of points with scatter 
import matplotlib.pyplot as plt

x_values = [1, 2, 3, 4, 5]
y_values = [1, 4, 9, 16, 25]   # square of each number 

plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=100)

# set chart title and label axes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

plt.show()
```

```python
# Calculating data automatically 
import matplotlib.pyplot as plt

x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, c='red', s=10)
 
# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Set the range for each axis.
ax.axis([0, 1100, 0, 1100000])   # 4 values: min and max values for the x-axis and the y-axis

plt.show()
```

```python
# # Defining Custom Colors
ax.scatter(x_values, y_values, c='red', s=10)   # c: color 

ax.scatter(x_values, y_values, c=(0, 0.8, 0), s=10)  # c: RGB color (0, 0.8, 0): light-green 
# # Values closer to 0 produce dark colors, and values closer to 1 produce lighter colors 
```

```python
# Using a colormap
import matplotlib.pyplot as plt

x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

plt.style.use('seaborn')
fig, ax = plt.subplots()
# cmap: low values a light color, high values a darker color 
ax.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Blues, s=10)
 
# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Set the range for each axis.
ax.axis([0, 1100, 0, 1100000]) 

plt.show()
```
